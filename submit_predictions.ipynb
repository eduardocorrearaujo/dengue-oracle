```python
import os
import pandas as pd
from itertools import product 
from dotenv import load_dotenv
from mosqlient import upload_prediction

# Load environment variables from .env file
load_dotenv()

# Access the environment variables
api_key = os.getenv('API_KEY')
```

```python
model_name = 'covar'
```

```python
if model_name == 'covar':
    model_id = 156

if model_name == 'lognorm':
    model_id = 155

model_id 
```

```python
states =  ['SC', 'PR', 'RS', 'SP', 'MG', 'RJ', 'ES', 'BA', 'CE', 'PE', 'PB', 'PI', 'RN', 
               'MA', 'AL', 'SE', 'DF', 'MT', 'MS', 'GO','RO', 'AC', 'AM', 'RR', 'PA', 'AP', 'TO']
    
for state, test_year in product(states, [2023, 2024,2025]): 
    
    df = pd.read_csv(f'predictions/preds_region_{model_name}_{state}_{test_year}.csv')

    if test_year == 2023: 
        val_test =1
    elif test_year == 2024:
        val_test =2 
    elif test_year == 2025:
        val_test = 3

    upload_prediction(
            model_id = model_id, 
            description = f'Validation test {val_test} for sprint 2025 preds in {state}', 
            commit = '08f57583e4bd75f6ed3e37818cfb958af7d9f250',
            predict_date = '2025-08-12', 
            prediction = df,
            adm_1=state,
            api_key = api_key)
```
