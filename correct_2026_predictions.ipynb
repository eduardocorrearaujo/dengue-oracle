Since the 2026 season has 53 weeks but the model only generates predictions for 52 weeks, we set the forecast for EW 202640 to be the same as for EW 202639.

```python
import pandas as pd 
from epiweeks import Week
from pathlib import Path

# Folder path
folder = Path("predictions/")

# Example: get all .csv files
files = [f.name for f in folder.iterdir() if f.is_file() and "2026" in f.name]

len(files)
```

```python
files[0]
```

```python
dates = pd.date_range(start = Week(2025, 41).startdate(), 
              end = Week(2026,40).startdate(),
              freq = 'W-SUN')
```

```python
for f in files: 
    df = pd.read_csv(f'predictions/{f}')
    
    df.loc[len(df)] = df.iloc[-1]
    
    df['date'] = dates

    df.to_csv(f'predictions/{f[:-4]}_corrected.csv', index = False)
    
```
