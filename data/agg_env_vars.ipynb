```python
import pandas as pd
```

```python
map = pd.read_csv('/Users/eduardoaraujo/data_sprint_2025/map_regional_health.csv')

map.head()
```

```python
df = pd.read_csv('/Users/eduardoaraujo/data_sprint_2025/environ_vars.csv.gz')

df = df.merge(map[['geocode', 'regional_geocode']])
df.head()
```

```python
df.loc[df.regional_geocode == 31095].biome.value_counts()/df.loc[df.regional_geocode ==31095].shape[0]
```

```python
df.groupby('regional_geocode')[['biome']].nunique().biome.value_counts()
```

```python
biome_counts = df.groupby(['regional_geocode', 'biome'])['geocode'].count().reset_index(name='count')

# Selecionar o bioma com maior contagem em cada regional_geocode
predominant_biome = biome_counts.sort_values('count', ascending=False).drop_duplicates('regional_geocode')

# Resultado final: apenas regional_geocode e bioma predominante
result = predominant_biome[['regional_geocode', 'biome']].reset_index(drop=True)
result['biome'] = result['biome'].replace(
    {'Caatinga':1.0, 'Mata Atlântica':2.0, 'Cerrado':3.0, 'Amazônia':4.0, 'Pampa':5.0,
       'Pantanal':6.0}
)

result
```

```python
result.to_csv('regional_biome.csv.gz', index = False)
```
