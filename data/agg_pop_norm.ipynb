```python
df_map = pd.read_csv('/Users/eduardoaraujo/Documents/sprint_data_2025/map_regional_health.csv')

df_map.head()
```

```python
df_map.groupby(['macroregion_name'])[['regional_geocode']].nunique()
```

```python
df_pop = pd.read_csv('/Users/eduardoaraujo/Documents/sprint_data_2025/datasus_population_2001_2024.csv.gz')
df_pop = df_pop.loc[df_pop.year==2022]        
df_pop.head()
```

```python
df_agg_pop = df_pop[['geocode', 'population']].merge(df_map[['regional_geocode', 'geocode']])

df_agg_pop = df_agg_pop.groupby(['regional_geocode'])[['population']].sum().reset_index()

df_agg_pop = df_agg_pop.rename(columns = {'population':'pop'})

df_agg_pop = df_agg_pop.merge(df_map[['regional_geocode', 'macroregion_name']].drop_duplicates())

df_agg_pop = df_agg_pop.merge(df_agg_pop.groupby(['macroregion_name'])[['pop']].max().reset_index().rename(columns = {
    'pop': 'pop_max'
}))

df_agg_pop['pop_norm'] =  df_agg_pop['pop']/df_agg_pop['pop_max']

df_agg_pop.head()
```
